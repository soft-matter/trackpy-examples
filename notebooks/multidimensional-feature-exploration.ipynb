## Exploring Feature Characterization in Multidimensional "Linked Brushing" Plots

The bulk of the labor in particle tracking is discovering a set of parameters that captures all the desired features and does not capture any spurious ones. In videos with many particles or videos with a "messy" background common in biological contexts, this problem in tricky.

The [particle-tracking tutorial by Eric Weeks](http://www.physics.emory.edu/faculty/weeks//idl/tracking.html) plots particles by "mass" (total brightness) and "size" (radius of gyration of brightness) and filters would-be features by their location in mass-size space. Trackpy provides a convenience function for making such a plot from a DataFrame of features, `trackpy.mass_size(f)`. We also find "signal" (peak brightness relative to the background) and eccentricity (0=circular, 1=linear) to be helpful indicators.

In the plot below, you can explore where clusters of true or spurious features might lie in this four-dimensional space of feature characterizations. We implement a four-by-four array of axes, plotting the characterization dimensions against each other. Further, if you click and drag to select a subset of the points on one set of axes, the points on all the axes are recolored to show their correspondence.

Try it. The plot below is interactive, even if you are reading this in a static notebook viewer.

### Prelimaries: Locate features in a sample image.

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

import trackpy as tp
```

```python
def gray(image):
    return image[:, :, 0]

frames = tp.ImageSequence('../sample_data/bulk_water/*.png', process_func=gray)
```

```python
f = tp.locate(frames[0], 11, invert=True)
```

### The Plot

This relies on the (relatively new) Python module [mpld3](https://github.com/jakevdp/mpld3). You can install it by typing

    pip install mpld3
    
into a Terminal or Command Prompt.

```python
import mpld3
from mpld3 import plugins, utils
mpld3.enable_notebook()

def linked_axes(df):
    "Draw an array of axes with linked brushing."
    L = df.shape[1]
    fig, ax = plt.subplots(L, L, sharex="col", sharey="row", figsize=(2*L, 2*L))
    fig.subplots_adjust(left=0.05, right=0.95, bottom=0.05, top=0.95,
                        hspace=0.1, wspace=0.1)

    for i in range(L):
        for j in range(L):
            points = ax[L - 1 - i, j].scatter(df.iloc[:, j], df.iloc[:, i],
                                              c='k', s=40, alpha=0.2)
            if i == 0:
                ax[L - 1 - i, j].set_xlabel(df.columns[j])
            if j == 0:
                ax[L - 1 - i, j].set_ylabel(df.columns[i])

    # remove tick labels
    for axi in ax.flat:
        for axis in [axi.xaxis, axi.yaxis]:
            axis.set_major_formatter(plt.NullFormatter())

    # Here we connect the linked brush plugin
    plugins.connect(fig, plugins.LinkedBrush(points))
    return fig
    
fig = linked_axes(f[['mass', 'size', 'ecc', 'signal']])
mpld3.display(fig)
```
